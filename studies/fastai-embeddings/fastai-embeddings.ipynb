## What is `trunc_normal_`?

```python
a = torch.empty(1, 10000, dtype=torch.float)
a.normal_()
plt.hist(a, 200, (-4.2, 4.2));
```

```python
a.fmod_(2)
plt.hist(a, 200, (-4.2, 4.2));
```

```python
a.mul_(0.01)
plt.hist(a, 200, (-.05, .05))
```

```python
a.add_(0)
plt.hist(a, 200, (-0.05, 0.05))
```
